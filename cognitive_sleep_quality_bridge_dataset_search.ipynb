## Conservative decision

The strongest **memory** candidate is the consumer-wearable working-memory dataset, because it directly names subjective sleep self-assessment, REM sleep, and N-back performance. It should be the first data-access target.

The strongest **reaction-time / processing-speed** candidate is the Oura + PVT + typing study, because it directly names sleep latency/restfulness and PVT reaction-test data. It should be requested from the authors.

MAUS is a useful backup for **PSQI → N-back**, but it is less ideal because PSQI is a global questionnaire and the dataset is mainly a workload-sensing dataset.

No installation formula is produced yet because no raw candidate dataset was downloaded and audited in this repository.
